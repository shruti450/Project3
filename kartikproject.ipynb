# Loading and Understanding the Data

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import warnings
warnings.filterwarnings('ignore')

# Load the dataset
file_path = '/kaggle/input/internet-search-engine-worldwide/search_engine_data.csv'
df = pd.read_csv('search_engine_data.csv')

# Display the first few rows of the dataframe
df.head()
```

```python
# Convert 'Date' column to datetime
df['Date'] = pd.to_datetime(df['Date'])

```

# Exploratory Data Analysis

```python
# Plotting the trend of search engine usage over time
plt.figure(figsize=(14, 8))
for column in df.columns[1:]:
    plt.plot(df['Date'], df[column], label=column)
plt.title('Search Engine Usage Over Time')
plt.xlabel('Date')
plt.ylabel('Usage')
plt.legend()
plt.show()
```

# Correlation Analysis

```python
# Select only numeric columns for correlation
numeric_df = df.select_dtypes(include=[np.number])

# Compute the correlation matrix
corr = numeric_df.corr()

# Plot the heatmap
plt.figure(figsize=(16, 12))
sns.heatmap(corr, annot=True, fmt='.2f', cmap='coolwarm')
plt.title('Correlation Heatmap of Search Engines')
plt.show()
```
